```python
url= 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents'

res_git = requests.get(url, headers=headers)

data = res_git.json()

list=[]

for i in range(len(data)):
    data = res_git.json()[i]['path']
    list.append(data)
list.pop(0)
list
```

```python
list_path=[]
contenido= []

for item in list:

    file_path=item

    url2= f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{file_path}'


    res_git = requests.get(url2, headers=headers)

    data = res_git.json()



    for i in range(len(data)):
        data = res_git.json()[i]['name']
        if 'scavengerhunt' in data:
            list_path.append(data)
            url3=f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{file_path}/{data}'
            print(url3)


            res_git3 = requests.get(url3, headers=headers)
            frase = res_git3.json()
            file_content = frase['content']
            file_content_encoding = frase.get('encoding')
            if file_content_encoding == 'base64':
                file_content = base64.b64decode(file_content).decode()
            contenido.append(file_content)
            print(contenido)
            print(file_content)

list_path
contenido
```

```python
list_files=list_path
resultado=contenido
```

```python
dictionary = dict(zip(list_files, resultado))
```

```python
sorted_keys = sorted(dictionary.keys())
sorted_dict = {key:dictionary[key] for key in sorted_keys}

frase_final=''

for key, value  in sorted_dict.items(): 
    frase_final +=value

frase_final.replace('\n', ' ')
```
